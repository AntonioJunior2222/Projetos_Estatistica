### C) Quais as cinco cidades brasileiras onde mais ocorreram acidentes em rodovias federais?


```python
municipios2020 = arq2020r['MUNICIPÍO'].value_counts().to_frame().head()
municipios2019 = arq2019r['MUNICIPÍO'].value_counts().to_frame().head()
municipios2018 = arq2018r['MUNICIPÍO'].value_counts().to_frame().head()
municipiostotal = arqtotal['MUNICIPÍO'].value_counts().to_frame().head()
municipiosgeral = pd.concat([municipios2020, municipios2019, municipios2018, municipiostotal], 
                            keys = ['2020', '2019', '2018', 'GERAL'])
municipiosgeral
```

Como nós queremos especificamente saber quais foram as cidades, eu filtrei apenas a coluna dos municípios, além disso usei "head()", já que nós queremos apenas as cinco cidades com maiores ocorrências. Após isso usei "concat" para agrupar os resultados em um único dataframe, e usei a função "Keys" para separar os dados por ano.

## 3º Para cada indicação abaixo, construa um novo dataframe, salvando-o com a terminação .csv.

### A) Ranking de acidentes por estado para todos os anos (um só arquivo combinando a informação dos três anos);


```python
a = arq2018r['ESTADO'].value_counts().to_frame().reset_index().rename(columns = {'index': 'ESTADO' ,'ESTADO': 'Nº OCORRÊNCIAS'})
b = arq2019r['ESTADO'].value_counts().to_frame().reset_index().rename(columns = {'index': 'ESTADO' ,'ESTADO': 'Nº OCORRÊNCIAS'})
c = arq2020r['ESTADO'].value_counts().to_frame().reset_index().rename(columns = {'index': 'ESTADO' ,'ESTADO': 'Nº OCORRÊNCIAS'})
d = arqtotal['ESTADO'].value_counts().to_frame().reset_index().rename(columns = {'index': 'ESTADO' ,'ESTADO': 'Nº OCORRÊNCIAS'})
rankinggeral = pd.concat([a, b, c, d], keys = ['2018', '2019', '2020', 'GERAL'],axis = 1)
print('RANKING GERAL DE OCORRÊNCIAS EM TODOS OS ANOS')
rankinggeral
```

```python
rankinggeral.to_csv('rankinggeral.csv', index = True)
#Para executar use:
# pd.read_csv('rankinggeral.csv')
```

#### COMENTÁRIOS: Para fazer esse ranking eu usei a função,  "value_counts()", utilizada anteriormente (2º letra C). Após isso concatenei os dataframes horizontalmente, usei concat, pois por ser um ranking era importante que os nomes dos estados se repetissem, já que eles mudam de posição ao longo dos anos e nos resultados gerais. Para converter em "CSV" usei a função "to_csv" e mantive o índice.

## B)  Acidentes por dia da semana para todos os anos (um só arquivo combinando a informação dos três anos).

```python
a = arq2018r['DIA DA SEMANA'].value_counts().to_frame()
b = arq2019r['DIA DA SEMANA'].value_counts().to_frame()
c = arq2020r['DIA DA SEMANA'].value_counts().to_frame()
d = arqtotal['DIA DA SEMANA'].value_counts().to_frame()
acidentessemana = pd.concat([a,b,c,d],axis=1)
acidentessemana.columns = ['2018', '2019', '2020', 'TOTAL']
acidentessemana
```

```python
acidentessemana.to_csv('acidentessemana.csv', index = True)
#Para executar use:
# pd.read_csv('acidentessemana.csv')
```

#### COMENTÁRIOS:  Nesta situação não era necessário fazer um ranking e sim a quantidade de acidentes nos dias da semana, eu apenas fiz concat sem a necessidade de repetir os dias da semana. Após isso usei o mesmo método da questão anterior para criar um arquivo "CSV".
